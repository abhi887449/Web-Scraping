### Loading data using request method and parsing html using beautifulsoup

```python
# importing beautiful soup
from bs4 import BeautifulSoup
# importing requests
import requests 
# importing pandas
import pandas as pd
```

```python
# storing response from https://www.amazon.in/
response = requests.get('https://www.amazon.in/')
print(response)
```

```python
response.headers
```

```python
data = response.text
# printing data
print(data)
```

```python
# creating beautifulsoup object
soup = BeautifulSoup(data,"html.parser")
print(soup.prettify())
```

```python
imgs = soup.find_all("img")
for i in imgs:
    print(i.get("src"))
```
